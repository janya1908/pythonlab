```python
# Tic-Tac-Toe Game in Python

def print_board(board):
    """Displays the current state of the Tic-Tac-Toe board."""
    print("\n")
    # Loop through board in increments of 3 to display 3 rows
    for i in range(0, 9, 3):
        # Print each row in the format: X | O | X
        print(f"{board[i]} | {board[i+1]} | {board[i+2]}")
        # Print a separator line between rows
        if i < 6:
            print("---------")

def check_winner(board, player):
    """Checks if the specified player has won the game."""
    # Define all possible winning combinations: rows, columns, and diagonals
    win_combinations = [
        [0, 1, 2], [3, 4, 5], [6, 7, 8], # Rows
        [0, 3, 6], [1, 4, 7], [2, 5, 8], # Columns
        [0, 4, 8], [2, 4, 6]             # Diagonals
    ]
    # Check if any winning combination is filled by the current player
    for combo in win_combinations:
        # If all spots in a combination are the same as the player's symbol, they win
        if all(board[i] == player for i in combo):
            return True
    return False  # No winning combination found

def check_tie(board):
    """Checks if the game is a tie (i.e., all positions are filled)."""
    # Return True if there are no empty spaces left on the board
    return all(spot != " " for spot in board)

def get_player_input(player, board):
    """Prompts the player for a valid move and returns the position index."""
    while True:
        try:
            # Get position from the player, adjusting for 0-based indexing
            position = int(input(f"Player {player}, enter a position (1-9): ")) - 1
            # Check if position is within the board and is unoccupied
            if position in range(9) and board[position] == " ":
                return position  # Return valid position
            else:
                print("Invalid position, try again.")  # Prompt for re-entry if invalid
        except ValueError:
            print("Please enter a valid number.")  # Handle non-integer inputs

def play_game():
    """Main function to control the flow of the Tic-Tac-Toe game."""
    # Initialize the game board with 9 empty spaces
    board = [" "] * 9
    # Set the starting player to "X"
    current_player = "X"

    # Game loop
    while True:
        # Display the current state of the board
        print_board(board)
        # Get the player's chosen position
        position = get_player_input(current_player, board)
        # Update the board with the current player's symbol at the chosen position
        board[position] = current_player

        # Check if the current player has won after the move
        if check_winner(board, current_player):
            print_board(board)  # Display final board state
            print(f"Player {current_player} wins!")  # Congratulate the winner
            break  # Exit the game loop

        # Check if the game is a tie after the move
        elif check_tie(board):
            print_board(board)  # Display final board state
            print("It's a tie!")  # Announce the tie
            break  # Exit the game loop

        # Switch to the other player for the next turn
        current_player = "O" if current_player == "X" else "X"

    # Offer a replay option at the end of the game
    if input("Play again? (y/n): ").lower() == "y":
        play_game()  # Restart the game if players choose "y"

# Start the game
play_game()

```

```python
# To-Do List Application in Python

def add_task(tasks):
    task = input("Enter a new task: ")
    tasks.append(task)
    print("Task added.")

def view_tasks(tasks):
    if tasks:
        print("\nCurrent Tasks:")
        for index, task in enumerate(tasks):
            print(f"{index + 1}. {task}")
    else:
        print("No tasks to display.")

def delete_task(tasks):
    try:
        view_tasks(tasks)
        index = int(input("Enter the task number to delete: ")) - 1
        if 0 <= index < len(tasks):
            removed_task = tasks.pop(index)
            print(f"Task '{removed_task}' removed.")
        else:
            print("Invalid task number.")
    except ValueError:
        print("Please enter a valid number.")

def main():
    tasks = []
    while True:
        print("\n1. Add Task\n2. View Tasks\n3. Delete Task\n4. Exit")
        choice = input("Choose an option: ")
        if choice == "1":
            add_task(tasks)
        elif choice == "2":
            view_tasks(tasks)
        elif choice == "3":
            delete_task(tasks)
        elif choice == "4":
            print("Exiting the to-do list application.")
            break
        else:
            print("Invalid choice. Try again.")

main()
```

```python
#ROBOT PATH PLANNING APPLICATION
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from queue import PriorityQueue
import matplotlib.colors as mcolors

def heuristic(a, b):
    """Calculate the Manhattan distance heuristic."""
    return abs(a[0] - b[0]) + abs(a[1] - b[1])

def a_star(grid, start, goal):
    """Perform A* pathfinding from start to goal."""
    rows, cols = grid.shape
    open_set = PriorityQueue()
    open_set.put((0, start))
    came_from = {}
    g_score = {start: 0}
    f_score = {start: heuristic(start, goal)}

    while not open_set.empty():
        _, current = open_set.get()

        if current == goal:
            # Reconstruct path
            path = []
            while current in came_from:
                path.append(current)
                current = came_from[current]
            path.append(start)
            return path[::-1]  # Reverse path from start to goal

        # Explore neighbors (up, down, left, right)
        for dx, dy in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
            neighbor = (current[0] + dx, current[1] + dy)
            if (0 <= neighbor[0] < rows and
                0 <= neighbor[1] < cols and
                grid[neighbor] == 0):  # Check if the neighbor is free
                tentative_g_score = g_score[current] + 1
                if neighbor not in g_score or tentative_g_score < g_score[neighbor]:
                    came_from[neighbor] = current
                    g_score[neighbor] = tentative_g_score
                    f_score[neighbor] = tentative_g_score + heuristic(neighbor, goal)
                    open_set.put((f_score[neighbor], neighbor))

    print("No path found")
    return None

def visualize(grid, path, start, goal):
    """Visualize the grid with the path, start, goal, and obstacles in color."""
    grid_display = np.zeros_like(grid, dtype=float)

    # Set colors: 0=free, 1=obstacle, 2=path, 3=start, 4=goal
    for position in path:
        grid_display[position] = 2  # Path as red

    grid_display[start] = 3       # Start as green
    grid_display[goal] = 4        # Goal as blue
    grid_display[grid == 1] = 1   # Obstacles as black

    # Define a custom colormap
    cmap = mcolors.ListedColormap(['white', 'black', 'red', 'green', 'blue'])
    bounds = [0, 1, 2, 3, 4]
    norm = mcolors.BoundaryNorm(bounds, cmap.N)

    # Display the grid with the custom color map
    plt.imshow(grid_display, cmap=cmap, norm=norm)
    plt.colorbar(ticks=[0, 1, 2, 3, 4], label='Legend: 0=Free, 1=Obstacle, 2=Path, 3=Start, 4=Goal')
    plt.show()

def main():
    # Input grid size
    rows, cols = map(int, input("Enter grid size (rows cols): ").split())
    grid = np.zeros((rows, cols), dtype=int)

    # Input obstacles
    num_obstacles = int(input("Enter number of obstacles: "))
    for _ in range(num_obstacles):
        x, y = map(int, input("Enter obstacle coordinates (row col): ").split())
        grid[x, y] = 1  # Mark obstacle with 1

    # Input start and goal positions
    start = tuple(map(int, input("Enter start position (row col): ").split()))
    goal = tuple(map(int, input("Enter goal position (row col): ").split()))

    # Display grid with pandas before pathfinding
    print("\nGrid Data:")
    print(pd.DataFrame(grid))

    # Find path
    path = a_star(grid, start, goal)
    if path:
        print("Path found:", path)
        visualize(grid, path, start, goal)  # Visualize path on the grid
    else:
        print("No valid path found.")

    # Option to retry with new start and goal positions
    retry = input("Try again with different start/goal? (y/n): ").lower()
    if retry == 'y':
        main()

if __name__ == "__main__":
    main()
```
