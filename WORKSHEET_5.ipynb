```python
#SOL 3
import numpy as np
from scipy import linalg

matrix = np.array([[3, 2, 1], [1, 1, 2], [2, 3, 1]])

determinant = linalg.det(matrix)

inverse = linalg.inv(matrix)

eigenvalues, _ = linalg.eig(matrix)

print(f"Matrix:\n{matrix}")
print(f"Determinant: {determinant}")
print(f"Inverse:\n{inverse}")
print(f"Eigenvalues: {eigenvalues}")
```

```python
#SOL 4
import numpy as np
from scipy import interpolate
import matplotlib.pyplot as plt

x = np.arange(0, 10)
y = np.sin(x)

f = interpolate.interp1d(x, y, kind='cubic')

x_new = np.linspace(0, 9, 100)
y_new = f(x_new)

plt.plot(x, y, 'o', label='Original data')
plt.plot(x_new, y_new, '-', label='Interpolated data')
plt.legend()
plt.show()
```

```python
#SOL 5
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

np.random.seed(0)
t = np.linspace(0, 1, 500)
signal_data = np.sin(2 * np.pi * 5 * t) + 0.5 * np.random.randn(t.size)

b, a = signal.butter(3, 0.1)
filtered_signal = signal.filtfilt(b, a, signal_data)

plt.plot(t, signal_data, label='Original Signal')
plt.plot(t, filtered_signal, label='Filtered Signal', linewidth=2)
plt.legend()
plt.show()
```

```python
#SOL 6
import numpy as np

np.random.seed(42)
sales_data = np.random.randint(100, 1000, (12, 4))

total_sales = np.sum(sales_data, axis=1)
average_sales = np.mean(sales_data, axis=1)
max_sales = np.max(sales_data, axis=1)

best_month = np.argmax(total_sales) + 1
worst_month = np.argmin(total_sales) + 1

print(f"Total Sales:\n{total_sales}\nAverage Sales:\n{average_sales}")
print(f"Best performing month: Month {best_month}")
print(f"Worst performing month: Month {worst_month}")
```

```python
#SOL 7
import pandas as pd

data = {'Student': ['Arin', 'Aditya', 'Chirag', 'Gurleen', 'Kunal'],
        'Math': [85, 79, 90, 66, 70],
        'Physics': [78, 82, 85, 75, 68],
        'Chemistry': [92, 74, 89, 80, 75],
        'English': [88, 90, 92, 78, 85]}

df_marks = pd.DataFrame(data)

df_marks['Total Marks'] = df_marks[['Math', 'Physics', 'Chemistry', 'English']].sum(axis=1)
df_marks['Average Marks'] = df_marks[['Math', 'Physics', 'Chemistry', 'English']].mean(axis=1)

top_performer = df_marks['Total Marks'].idxmax()
bottom_performer = df_marks['Total Marks'].idxmin()

passing_percentage = (df_marks[['Math', 'Physics', 'Chemistry', 'English']] >= 40).mean().mean() * 100

print(df_marks)
print(f"Top Performer: {df_marks.iloc[top_performer]['Student']}")
print(f"Bottom Performer: {df_marks.iloc[bottom_performer]['Student']}")
print(f"Passing Percentage: {passing_percentage:.2f}%")
```

```python
#SOL 8
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

time = np.array([0, 1, 2, 3, 4, 5])
velocity = np.array([2, 3.1, 7.9, 18.2, 34.3, 56.2])

def quadratic(t, a, b, c):
    return a * t**2 + b * t + c

params, _ = curve_fit(quadratic, time, velocity)
a, b, c = params

fitted_velocity = quadratic(time, a, b, c)

plt.plot(time, velocity, 'o', label='Original Data')
plt.plot(time, fitted_velocity, '-', label='Fitted Data')
plt.legend()
plt.show()

print(f"Fitted parameters: a = {a}, b = {b}, c = {c}")
```
