```python
#SOL 1
import matplotlib.pyplot as plt

students = ['Arin', 'Aditya', 'Chirag', 'Gurleen', 'Kunal']
math = [85, 79, 90, 66, 70]
physics = [78, 82, 85, 75, 68]
chemistry = [92, 74, 89, 80, 75]
english = [88, 90, 92, 78, 85]

total_marks = [math[i] + physics[i] + chemistry[i] + english[i] for i in range(len(students))]
average_marks = [total / 4 for total in total_marks]

top_performer = students[total_marks.index(max(total_marks))]
bottom_performer = students[total_marks.index(min(total_marks))]

math_avg = sum(math) / len(math)
physics_avg = sum(physics) / len(physics)
chemistry_avg = sum(chemistry) / len(chemistry)
english_avg = sum(english) / len(english)

pass_fail = [['Pass' if subject >= 40 else 'Fail' for subject in [math[i], physics[i], chemistry[i], english[i]]] for i in range(len(students))]

plt.bar(students, total_marks, color='skyblue')
plt.title('Student Performance - Total Marks')
plt.xlabel('Students')
plt.ylabel('Total Marks')
plt.show()

print(f"Top Performer: {top_performer}")
print(f"Bottom Performer: {bottom_performer}")
print(f"Subject Averages: Math = {math_avg:.2f}, Physics = {physics_avg:.2f}, Chemistry = {chemistry_avg:.2f}, English = {english_avg:.2f}")
print(f"Pass/Fail Status: {pass_fail}")
```

```python
#SOL 2
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

time = np.array([0, 1, 2, 3, 4, 5])
velocity = np.array([2, 3.1, 7.9, 18.2, 34.3, 56.2])

def quadratic(t, a, b, c):
    return a*t**2 + b*t + c

params, _ = curve_fit(quadratic, time, velocity)
a, b, c = params

plt.scatter(time, velocity, color='red', label='Data')
t_vals = np.linspace(0, 5, 100)
plt.plot(t_vals, quadratic(t_vals, a, b, c), label=f'Fit: {a:.2f}t² + {b:.2f}t + {c:.2f}')
plt.title('Velocity vs Time with Quadratic Fit')
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.legend()
plt.show()
```

```python
#SOL 3
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

year = [2000, 2005, 2010, 2015, 2020]
population = [50, 55, 70, 80, 90]

correlation, _ = stats.pearsonr(year, population)
print(f"Pearson's Correlation Coefficient: {correlation}")

from scipy.interpolate import interp1d

f_interp = interp1d(year, population)
pop_2008 = f_interp(2008)
print(f"Estimated population in 2008: {pop_2008}")

plt.scatter(year, population, color='blue', label='Data')
plt.plot(year, f_interp(year), color='red', label='Linear Interpolation')
plt.scatter(2008, pop_2008, color='green', label='Estimated (2008)')
plt.title('Population Over Years')
plt.xlabel('Year')
plt.ylabel('Population (in thousands)')
plt.legend()
plt.show()
```

```python
#SOL 4
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt

def p(x):
    return 3*x**3 - 5*x**2 + 2*x - 8

roots = np.roots([3, -5, 2, -8])
print(f"Roots of the polynomial: {roots}")

x_vals = np.linspace(-3, 3, 400)
y_vals = p(x_vals)

plt.plot(x_vals, y_vals, label='p(x) = 3x³ - 5x² + 2x - 8')
plt.axhline(0, color='black',linewidth=1)
plt.scatter(roots, p(roots), color='red', label='Roots')
plt.title('Polynomial and Roots')
plt.xlabel('x')
plt.ylabel('p(x)')
plt.legend()
plt.show()
```

```python
#SOL 5
import time
import string
import random

def generate_text_file(size_mb, file_name):
    size = size_mb * 1024 * 1024
    chars = ''.join(random.choices(string.ascii_lowercase + string.whitespace, k=size))
    with open(file_name, 'w') as f:
        f.write(chars)

def convert_to_upper(file_name):
    start = time.time()
    with open(file_name, 'r') as f:
        text = f.read()
    text_upper = text.upper()
    end = time.time()
    return end - start


file_sizes = [200, 400, 600, 800, 1000]
times = []
for size in file_sizes:
    file_name = f'text_{size}MB.txt'
    generate_text_file(size, file_name)
    duration = convert_to_upper(file_name)
    times.append(duration)
    print(f"Time taken to convert {size}MB file: {duration:.2f} seconds")

plt.plot(file_sizes, times, marker='o')
plt.title('Time taken to convert text files to upper case')
plt.xlabel('File Size (MB)')
plt.ylabel('Time (seconds)')
plt.show()
```

```python
#SOL 6
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

def f(x):
    return x*4 - 3*x*3 + 2

initial_guess = 0
result = minimize(f, initial_guess)


local_min_x = result.x[0]
local_min_y = f(local_min_x)

x_vals = np.linspace(-2, 3, 500)
y_vals = f(x_vals)

plt.plot(x_vals, y_vals, label=r'$f(x) = x^4 - 3x^3 + 2$', color='b')

plt.scatter(local_min_x, local_min_y, color='r', zorder=5, label=f'Local Minima at x = {local_min_x:.2f}')

plt.title('Function Plot and Local Minima')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()

plt.grid(True)
plt.show()

print(f"Local minima is at x = {local_min_x:.4f}, f(x) = {local_min_y:.4f}")
```
